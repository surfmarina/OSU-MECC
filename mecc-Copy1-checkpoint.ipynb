# Sebastian- Generating Spectrums From Signigifigant Wave Height

```python
# JONSWAP AVERAGE SEA STATE (FLORIDA)
f = np.linspace(0.01,1,500)
Hs = 0.18
Tp = 2.50
S1 = wave.resource.jonswap_spectrum(f, Tp, Hs, gamma=3.3) 
S1.plot(xlabel='frequency [Hz]', ylabel = 'response[m^2/Hz]')
```

```python
S1.to_csv(r'C:\Users\storm\Box\MECC Testing Conditions\Oregon Testing\Scaled Data\WinterSpectrum.csv', index = False)
```

```python
# JONSWAP spectrum for creating short-term time-series - elevation profile for 20 minutes
# numpy.linspace(start, stop, num=50

t_st = 10.0 * 60 # 10 minutes
# time in seconds
time = np.linspace(0, t_st, 600000)#2*Nf+1)
eta = wave.resource.surface_elevation(S1,time)
plt.plot(time,eta)
print(len(eta))
eta.to_csv(r'C:\Users\storm\Box\MECC Testing Conditions\Florida Testing\Scaled Data\WinterFreude.csv', index = False)
```
